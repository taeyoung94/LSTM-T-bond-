```python
def ParsePrice():
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'lxml')
    price = soup.find('span', class_ = 'Trsdu(0.3s) Fw(b) Fz(36px) Mb(-4px) D(ib)').text
    return price
```

```python
while True:
    print('Stock Current Price is: ' + str(ParsePrice()))
```
